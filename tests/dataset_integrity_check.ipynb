```python
def check_neighbour(data_path, source_idx=3):
    # file_pattern = os.path.join(data_path, 'data_*.npy')
    # files = glob.glob(file_pattern)
    # for file in files:
    data_set = UM2N.MeshDataset(data_path)
    for i in range(len(data_set)):
        coords = data_set[i].x[:, :2]
        num_nodes = coords.shape[0]
        source_mask = index_to_mask(
            torch.tensor([source_idx]), num_nodes
        )
        nei = UM2N.get_neighbors(source_mask, data_set[i].edge_index)
        if (nei.sum() == 6):
            pass
        else:
            raise ValueError(f"In dataset {data_path} The number of neighbors is not 6")
    return


check_neighbour("/Users/cw1722/Documents/warpmesh/data/helmholtz/z=<0,1>_ndist=None_max_dist=6_<16x16>_n=100_iso_full/data")
```
